```python
from pyspark.context import SparkContext
from pyspark.sql.session import SparkSession
sc = SparkContext('local')
spark = SparkSession(sc)
```

```python
df = spark.read.json("/content/nyt2.json")
df.printSchema()
df.show()
```

```python
filepath = "/content/nyt2.json"
nyt = spark.read.json(filepath).rdd
nyt.take(5)
```

```python
nyt.first().published_date[0][0]
```

#  LIST OF BOOK FOR AUTHOR




```python
sasso = nyt.map(lambda x: (x.title, x.published_date[0][0]))
sasso.first()
```

```python
culo = nyt.map(lambda x: (x.author, x.title))
culo.collect()
```

```python
grouped = culo.groupByKey()
grouped = grouped.map(lambda x: (x[0], list(dict.fromkeys(x[1]))))
grouped.take(4)
```

```python
column = ["Author","Books"]
df = spark.createDataFrame(data = grouped, schema = column)
df.printSchema()
df.show(truncate=False)
```

```python
mas = grouped.map(lambda x : (x[0], len(x[1])))
mas.take(4)
```

```python
from pyspark.sql.functions import col
column = ["Author","Number of books published"]

df = spark.createDataFrame(data = mas, schema = column)
df.orderBy('Number of books published', ascending=False).show()
```

```python
sasso = nyt.map(lambda x: (x._id[0], x.published_date[0][0]))
sasso.first()
column= ["Time", "TimeUNIX"]
df=spark.createDataFrame(data = sasso, schema = column)
df.printSchema()
df.show(truncate=False)
```

```python
from pyspark.sql.functions import col
df3=df.select(col("TimeUNIX"),"MM-dd-yyyy").alias("TimeUNIX")
df3.printSchema()
df3.show(truncate=False)
```
